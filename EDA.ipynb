```python
import pandas as pd
from glob import glob
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

pd.options.display.max_columns = None

def dataframe_from_csv(target):
    return pd.read_csv(target).rename(columns=lambda x: x.strip())

def dataframe_from_csvs(targets):
    return pd.concat([dataframe_from_csv(x) for x in targets])
```

```python
TRAIN_DATASET = dataframe_from_csvs(sorted([x for x in Path("data/training/").glob("*.csv")]))
TRAIN_DATASET
```

```python
TEST_DATASET = dataframe_from_csvs(sorted([x for x in Path("data/testing/").glob("*.csv")]))
TEST_DATASET
```

```python
VALIDATION_DATASET = dataframe_from_csvs(sorted([x for x in Path("data/validation/").glob("*.csv")]))
VALIDATION_DATASET
```

```python
display(TRAIN_DATASET.describe())
display(VALIDATION_DATASET.describe())
display(VALIDATION_DATASET[VALIDATION_DATASET.attack == 0].describe())
display(VALIDATION_DATASET[VALIDATION_DATASET.attack==1].describe())
```

```python
TAG_MIN = TRAIN_DATASET[TRAIN_DATASET.columns.drop(['time'])].min()
TAG_MAX = TRAIN_DATASET[TRAIN_DATASET.columns.drop(['time'])].max()
```

```python
def normalize(df):
    ndf = df.copy()
    for c in df.columns:
        if TAG_MIN[c] == TAG_MAX[c]:
            ndf[c] = df[c] - TAG_MIN[c]
        else:
            ndf[c] = (df[c] - TAG_MIN[c]) / (TAG_MAX[c] - TAG_MIN[c])
    return ndf
```

```python
TRAIN_DF = normalize(TRAIN_DATASET[TRAIN_DATASET.columns.drop(['time'])]).ewm(alpha=0.9).mean()
VALIDATION_DF = normalize(VALIDATION_DATASET[VALIDATION_DATASET.columns.drop(['time', 'attack'])]).ewm(alpha=0.9).mean()
```

```python
display(TRAIN_DF)
display(VALIDATION_DF)
```

```python
def boundary_check(df):
    x = np.array(df, dtype=np.float32)
    return np.any(x > 1.0), np.any(x < 0), np.any(np.isnan(x))
```

```python
print(boundary_check(TRAIN_DF))
print(boundary_check(VALIDATION_DF))
```
